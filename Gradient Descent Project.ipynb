```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
```

```python
train_data=np.genfromtxt("0000000000002417_training_boston_x_y_train.csv",delimiter=',')
x_train=train_data[:,0:13]
y_train=train_data[:,13]

scaler=preprocessing.MinMaxScaler(feature_range=(0,1))#scaler object created
scaler.fit(x_train)#find those parameter on which x_train has been scaled
scaler.transform(x_train)
x_train = np.c_[x_train, np.ones(len(x_train))]
```

```python
def predict(x, m):
    #for each row ,we have coressponding one output
    y = np.zeros(len(x))
    for i in range(len(x)):
        y[i] = sm(x, m, i)
    return y
```

```python
#cost func(mean square error) = ((truth value-predict value)**2).mean()
def cost(x, y, m):
    X_M = np.matmul(x, m)
    return ((y - X_M) ** 2).mean()
```

```python
#sm= m1x1 + m2x2 +m3x3 +....+ mnxn
def sm(x, m, i):
    s = 0
    for k in range(14):
        s += m[k] * x[i][k]
    return s
```

```python
def step_grad(x, y, learning_rate, m):
    m_slope = 0
    M = len(x)
    for j in range(14):#for each column
        for i in range(M):#i will go through each row
            m_slope += (-2) * ((y[i] - sm(x, m, i)) * x[i][j]).mean()
        m[j] = m[j] - learning_rate * m_slope
    return m
```

```python
def gd(x, y, learning_rate, num_iter):
    m = np.zeros(14)
    for k in range(num_iter):#for each iteration value of each feature of m has been updated
        m = step_grad(x, y, learning_rate, m)#current value of each feature of m passes
        print(k, "Cost:", cost(x, y, m))
    return m
```

```python
def run (x_train, y_train):
    #learning rate =the rate at which value of m increases or decreases
    learning_rate = 0.00025
    num_iter = 2400
    m = gd(x_train, y_train, learning_rate, num_iter)
    x_test = np.genfromtxt("0000000000002417_test_boston_x_test.csv", delimiter=",")
    x_test = scaler.transform(x_test)
    x_test = np.c_[x_test, np.ones(len(x_test))]
    y_pred = predict(x_test, m)
    np.savetxt("Pred_GD_boston.csv", y_pred)
```

```python
run(x_train,y_train)
```
